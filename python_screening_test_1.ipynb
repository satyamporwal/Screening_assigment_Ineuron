Question 1: -
Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.
Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input - string = “write write write all the number from from from 1 to 100”
Example output - 5
Explanation - From the given string we can note that the most frequent words are “write” and “from” and
the maximum value of both the values is “write” and its corresponding length is 5

```python
from collections import Counter

def find_highest_frequency_word_length(string):
    words = string.split()
    word_counts = Counter(words)

    max_frequency = max(word_counts.values())
    highest_frequency_word_length = max(len(word) for word, count in word_counts.items() if count == max_frequency)

    return word_counts,highest_frequency_word_length
    
```

```python
from collections import Counter

def is_valid_string(s):
    s =  s.replace(' ','')
    # Count the frequency of each character in the string
    char_frequency = Counter(s)

    # Get the count of the most common character
    max_count = char_frequency.most_common(1)[0][1]

    # Count the number of characters with frequencies different from the most common
    count_diff = sum(1 for count in char_frequency.values() if count != max_count)

    # Check if the string is valid
    if count_diff == 0:
        # All characters have the same frequency
        return "YES"
    elif count_diff == 1:
        # Remove one character to make all frequencies the same
        return "YES"
    else:
        # More than one character needs to be removed
        return "NO"

# Example usage:
string1 = "my nae is"
result1 = is_valid_string(string1)
print(result1)  # Output: YES

string2 = "a b c d ee g h"
result2 = is_valid_string(string2)
print(result2)  # Output: NO
```

Excepted Output Data Attributes -

```python
import pandas as pd
import requests
import json
import datetime


def download_data(url):
    response = requests.get(url)
    data = response.json()
    return data


def extract_data(data):
    episodes = []

    for episode_data in data:
        episode = {}

        # Extracting the desired attributes from each episode data
        episode['id'] = episode_data['id']
        episode['url'] = episode_data['url']
        episode['name'] = episode_data['name']
        episode['season'] = episode_data['season']
        episode['number'] = episode_data['number']
        episode['type'] = episode_data['type']
        episode['airdate'] = datetime.datetime.strptime(episode_data['airdate'], "%Y-%m-%d").date()
        episode['airtime'] = datetime.datetime.strptime(episode_data['airtime'], "%H:%M").strftime("%I:%M %p")
        episode['runtime'] = float(episode_data['runtime'])
        episode['rating'] = float(episode_data['rating']['average'])
        episode['summary'] = episode_data['summary'].replace("<p>", "").replace("</p>", "")
        episode['medium_image'] = episode_data['image']['medium']
        episode['original_image'] = episode_data['image']['original']

        episodes.append(episode)

    return episodes


if __name__ == '__main__':
    url = 'http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes'

    # Downloading the data
    episode_data = download_data(url)

    # Extracting the desired data
    extracted_data = extract_data(episode_data['_embedded']['episodes'])

    # Creating a DataFrame
    df = pd.DataFrame(extracted_data)
    df.head()

    # Printing the DataFrame
   

```
